```python
import numpy as np
import matplotlib.pyplot as plt


def r_point_upper_half_grid(N):
    r_rel_x = []
    r_rel_y = []
    x_span = int((int(N))/2)+1
    r_pair_1 = 0
    for i in range(x_span):
        for j in range(x_span):
            if(i<=j):
              r_rel_x.append(i)
              r_rel_y.append(j)
              r_pair_1 = r_pair_1+1
    R_relative_1 = np.stack((r_rel_x,r_rel_y),axis = 1)
 
    print("no of r points", r_pair_1)
    return R_relative_1,r_pair_1


def r_point_full_grid(N):
    r_rel_x = []
    r_rel_y = []
    x_span = int((int(N))/2)+1
    r_pair_2 = 0
    for i in range(x_span):
        for j in range(x_span):
              r_rel_x.append(i)
              r_rel_y.append(j)
              r_pair_2 = r_pair_2+1
    R_relative_2 = np.stack((r_rel_x,r_rel_y),axis = 1)
    print("no of r points", r_pair_2)

    return R_relative_2,r_pair_2
```

```python
def r_space_full_grid_cons(r_x,r_y):
    
    
    R_00 = []
    R_lab = np.arange(len(r_x))
    R_lab_00 = []
    for i0 in range(len(r_x)):
        R_00.append((r_x[i0],r_y[i0]))
        R_lab_00.append(R_lab[i0])
    
    R_01 = []
    R_lab_01 = []
    
    for i1 in range(len(R_00)):
        r_y_0,r_x_0 = R_00[i1]
        
        if (r_x_0,r_y_0) not in R_00:
           R_01.append((r_x_0,r_y_0))
           R_lab_01.append(R_lab_00[i1])
    
    R_0 = R_00+R_01
    R_lab_0 = R_lab_00+R_lab_01

#========================================================================================================

    R_1 = []
    R_lab_1 = []
    for i2 in range(len(R_0)):
        
        r_x_temp,r_y_temp = R_0[i2]
        r_x_1 = -1*r_x_temp
        r_y_1 = r_y_temp
        
        
        if (r_x_1,r_y_1) not in R_0:
            
           R_1.append((r_x_1,r_y_1))
           R_lab_1.append(R_lab_0[i2])
           
    R_0_p_1 = R_0+R_1
    R_lab_0_p_1 = R_lab_0+R_lab_1
    
#======================================================================================================


    R_2 = []
    R_lab_2 = []
    
    for i3 in range(len(R_0)):
        r_x_temp,r_y_temp = R_0[i3]
        r_x_2 = -1*r_x_temp
        r_y_2 = -1*r_y_temp
        
        if (r_x_2,r_y_2) not in R_0_p_1:
            
           R_2.append((r_x_2,r_y_2))
           R_lab_2.append(R_lab_0[i3])
           
    R_0_p_1_p_2 = R_0_p_1+R_2
    R_lab_0_p_1_p_2 = R_lab_0_p_1+R_lab_2
    
#============================================================================================


    R_3 = []
    R_lab_3 = []
    
    for i4 in range(len(R_0)):
        r_x_temp,r_y_temp = R_0[i4]
        r_x_3 = r_x_temp
        r_y_3 = -1*r_y_temp
        
        if (r_x_3,r_y_3) not in R_0_p_1_p_2:
            
           R_3.append((r_x_3,r_y_3))
           R_lab_3.append(R_lab_0[i4])
           
    R_full = R_0_p_1_p_2+R_3
    R_lab_full = R_lab_0_p_1_p_2+R_lab_3
    
    R_sq_full = []
    R_x_full = []
    R_y_full = []
    R_label = []
    for i in range(len(R_full)):
        r_x_temp,r_y_temp = R_full[i]
        r_2 = r_x_temp*r_x_temp+r_y_temp*r_y_temp
        R_sq_full.append(r_2)
        R_x_full.append(r_x_temp)
        R_y_full.append(r_y_temp)
        R_label.append(R_lab_full[i])
     
    
    plt.figure()
    plt.scatter(R_x_full,R_y_full,c=R_label)
    return R_x_full,R_y_full,R_sq_full,R_label
```

```python
r_rel_uhp,rp_uhp = r_point_upper_half_grid(10)
```

```python
r_rel_fp,rp_fp = r_point_full_grid(10)
```

```python
r_x_full_0,r_y_full_0,r_2_full_0,r_l_0 = r_space_full_grid_cons(r_rel_uhp[:,0],r_rel_uhp[:,1])
```

```python
r_x_full_1,r_y_full_1,r_2_full_1,r_l_1 = r_space_full_grid_cons(r_rel_fp[:,0],r_rel_fp[:,1])
```
