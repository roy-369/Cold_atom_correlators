```python
def arr_loc(val,X):
   ind = 0
   for i in range(len(X)):
      if(round(float(X[i]),2) == round(val,2)):
         ind = i
         break
   return ind
```

```python
def kinetic_energy_half_filling_line_plots(Text_dir_en,Graph_dir,N,u_hf,dtau):
        
    filename_ke_0 = "%s/Kinetic_Energy_data_half_filling_data_N_%s_U_%s_dtau_%s.dat"%(Text_dir_en,N,u_hf[0],dtau)
    beta_val,ke = np.loadtxt(filename_ke_0,unpack = 'True',usecols = [0,1])
    
    U_val = np.zeros(len(u_hf))
    Beta_val = np.zeros((len(u_hf),len(beta_val)))
    KE = np.zeros((len(u_hf),len(beta_val)))
    KE_std = np.zeros((len(u_hf),len(beta_val)))
    dKE_dT = np.zeros((len(u_hf),len(beta_val)))
    dKE_dT_std = np.zeros((len(u_hf),len(beta_val)))
    Cv_KE =  np.zeros((len(u_hf),len(beta_val)))

    for i in range(len(u_hf)):
        U_val[i] = float(u_hf[i])
        filename_ke = "%s/Kinetic_Energy_data_half_filling_data_N_%s_U_%s_dtau_%s.dat"%(Text_dir_en,N,u_hf[i],dtau)
        Beta_val[i,:],KE[i,:],KE_std[i,:],dKE_dT[i,:],dKE_dT_std[i,:] = np.loadtxt(filename_ke,unpack = 'True',usecols = [0,1,2,3,4])
        Cv_KE[i,:]=np.gradient(KE[i,:],beta_val)
        
    print("KE",KE)    
    plt.figure(figsize = (25,20))
    plt.xticks(fontsize = 80)
    plt.yticks(fontsize = 80)
    for k in range(len(beta_val)):
        T = beta_val[k]
        plt.errorbar(U_val,KE[:,k],yerr=KE_std[:,k],marker = "o",markersize = 20,elinewidth=3,capsize=5,label = "T=%s"%(str(round(T,3))))
    plt.grid("True",which='both')
    plt.legend(loc='best',ncols =4,fontsize=60)
    #plt.xlim(0.75,1.25)
    plt.axhline(y = 0,color = 'black',linewidth = 2,linestyle = 'dashed')
    plt.tight_layout()
    plt.savefig("%s/Kinetic_energy_half_filling_N_%s_dtau_%s.png"%(Graph_dir,N,dtau))
    plt.close('all')


    plt.figure(figsize = (25,20))
    plt.xticks(fontsize = 80)
    plt.yticks(fontsize = 80)
    for k in range(len(beta_val)-6):
        T = beta_val[k]
        plt.errorbar(U_val,dKE_dT[:,k],yerr=dKE_dT_std[:,k],marker = "o",markersize = 20,elinewidth=3,capsize=5,label = "T=%s"%(str(round(T,3))))
    plt.grid("True",which='both')
    plt.legend(loc='best',ncols =3,fontsize=60)
    #plt.xlim(0.75,1.25)
    plt.axhline(y = 0,color = 'black',linewidth = 2,linestyle = 'dashed')
    plt.tight_layout()
    plt.savefig("%s/Temp_derivative_Kinetic_energy_half_filling_N_%s_dtau_%s.png"%(Graph_dir,N,dtau))
    plt.close('all')

    plt.figure(figsize = (25,20))
    plt.xticks(fontsize = 80)
    plt.yticks(fontsize = 80)
    for k in range(len(beta_val)-6):
        T = beta_val[k]
        plt.errorbar(U_val,Cv_KE[:,k],yerr=dKE_dT_std[:,k],marker = "o",markersize = 20,elinewidth=3,capsize=5,label = "T=%s"%(str(round(T,3))))
    plt.grid("True",which='both')
    plt.legend(loc='best',ncols =3,fontsize=60)
    #plt.xlim(0.75,1.25)
    plt.axhline(y = 0,color = 'black',linewidth = 2,linestyle = 'dashed')
    plt.tight_layout()
    plt.savefig("%s/Numerical_Temp_derivative_Kinetic_energy_half_filling_N_%s_dtau_%s.png"%(Graph_dir,N,dtau))
    plt.close('all')
    
```
