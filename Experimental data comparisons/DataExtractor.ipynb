```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data.csv')

# Get all available interaction values
U_vals = df['U_val'].unique()
print(f'Available U values: {U_vals}')

# Get available correlator types
types = [_c[:-4] if ((_c[0]=='c') & (_c[-3:]=='val')) else 'none' for _c in df.columns.values]
types = list(filter(('none').__ne__, types))
print(f'Correlation types: {types}')
```

```python
C_dh = np.zeros((4,15))
g_dh = np.zeros((4,15))
C_dd = np.zeros((4,15))
C_ds = np.zeros((4,15))
C_hh = np.zeros((4,15))
C_nd = np.zeros((4,15))
C_nh = np.zeros((4,15))
C_nn = np.zeros((4,15))
C_ns = np.zeros((4,15))
C_sh = np.zeros((4,15))
C_ss = np.zeros((4,15))


C_dh_err = np.zeros((4,15))
C_dd_err = np.zeros((4,15))
C_ds_err = np.zeros((4,15))
C_hh_err = np.zeros((4,15))
C_nd_err = np.zeros((4,15))
C_nh_err = np.zeros((4,15))
C_nn_err = np.zeros((4,15))
C_ns_err = np.zeros((4,15))
C_sh_err = np.zeros((4,15))
C_ss_err = np.zeros((4,15))


Num_den = np.zeros((4,15))
Num_dbn = np.zeros((4,15))
Num_hln = np.zeros((4,15))
Num_mmt = np.zeros((4,15))

Num_den_err = np.zeros((4,15))
Num_dbn_err = np.zeros((4,15))
Num_hln_err = np.zeros((4,15))
Num_mmt_err = np.zeros((4,15))


for i in range(len(U_vals)):

    u_val = U_vals[i]
    dx = 1; dy = 1;

    _df = df[df['U_val']==u_val]
    _df = _df[(_df['dx']==dx) & (_df['dy']==dy)]

    n_vals    = _df['n_val'   ].values; n_errs    = _df['n_err'   ].values
    nd_vals   = _df['nd_val'  ].values; nd_errs   = _df['nd_err'  ].values
    nh_vals   = _df['nh_val'  ].values; nh_errs   = _df['nh_err'  ].values
    c_dh_vals = _df['c_dh_val'].values; c_dh_errs = _df['c_dh_err'].values
    c_dd_vals = _df['c_dd_val'].values; c_dd_errs = _df['c_dd_err'].values
    c_ds_vals = _df['c_ds_val'].values; c_ds_errs = _df['c_ds_err'].values
    c_hh_vals = _df['c_hh_val'].values; c_hh_errs = _df['c_hh_err'].values
    c_nd_vals = _df['c_nd_val'].values; c_nd_errs = _df['c_nd_err'].values
    c_nh_vals = _df['c_nh_val'].values; c_nh_errs = _df['c_nh_err'].values
    c_nn_vals = _df['c_nn_val'].values; c_nn_errs = _df['c_nn_err'].values
    c_ns_vals = _df['c_ns_val'].values; c_ns_errs = _df['c_ns_err'].values
    c_sh_vals = _df['c_sh_val'].values; c_sh_errs = _df['c_sh_err'].values
    c_ss_vals = _df['c_ss_val'].values; c_ss_errs = _df['c_ss_err'].values

    print(u_val)
    print(c_dh_vals.shape,"c_dh shape")
    
    C_dh[i,:] = np.copy(np.asarray(c_dh_vals))
    C_dd[i,:] = np.copy(np.asarray(c_dd_vals))
    C_ds[i,:] = np.copy(np.asarray(c_ds_vals))
    C_hh[i,:] = np.copy(np.asarray(c_hh_vals))
    C_nd[i,:] = np.copy(np.asarray(c_nd_vals))
    C_nh[i,:] = np.copy(np.asarray(c_nh_vals))
    C_nn[i,:] = np.copy(np.asarray(c_nn_vals))
    C_ns[i,:] = np.copy(np.asarray(c_ns_vals))
    C_sh[i,:] = np.copy(np.asarray(c_sh_vals))
    C_ss[i,:] = np.copy(np.asarray(c_ss_vals))
    
    Num_den[i,:] = np.copy(np.asarray(n_vals))
    Num_dbn[i,:] = np.copy(np.asarray(nd_vals))
    Num_hln[i,:] = np.copy(np.asarray(nh_vals))
    Num_mmt[i,:] = np.subtract(np.ones(15),np.add(Num_dbn[i,:],Num_hln[i,:]))
    
    C_dh_err[i,:] = np.copy(np.asarray(c_dh_errs))
    C_dd_err[i,:] = np.copy(np.asarray(c_dd_errs))
    C_ds_err[i,:] = np.copy(np.asarray(c_ds_errs))
    C_hh_err[i,:] = np.copy(np.asarray(c_hh_errs))
    C_nd_err[i,:] = np.copy(np.asarray(c_nd_errs))
    C_nh_err[i,:] = np.copy(np.asarray(c_nh_errs))
    C_nn_err[i,:] = np.copy(np.asarray(c_nn_errs))
    C_ns_err[i,:] = np.copy(np.asarray(c_ns_errs))
    C_sh_err[i,:] = np.copy(np.asarray(c_sh_errs))
    C_ss_err[i,:] = np.copy(np.asarray(c_ss_errs))
    
    Num_den_err[i,:] = np.copy(np.asarray(n_errs))
    Num_dbn_err[i,:] = np.copy(np.asarray(nd_errs))
    Num_hln_err[i,:] = np.copy(np.asarray(nh_errs))
    Num_mmt_err[i,:] = np.sqrt(np.add(np.power(Num_dbn_err[i,:],2),np.power(Num_hln_err[i,:],2)))
    
```

```python

plt.figure()

for i in range(len(U_vals)):
    for j in range(15):
        g_dh[i,j] = C_dh[i,j]/(Num_dbn[i][j]*Num_hln[i,j]+0.0000001)
        
for i in range(2): #len(U_vals)):   
    plt.plot(Num_den[i,:-2],g_dh[i,:-2],marker = "o",label="U=%s"%str(U_vals[i]))
    plt.legend(loc= 'best')
```

```python
plt.figure()
plt.errorbar(Num_den[0,:],C_dh[0,:],yerr = C_dh_err[0,:],marker = 'o', elinewidth = 3, capsize = 5,label = "C_mm, U = 7.1")
#plt.plot(Num_den[0,:],np.add(C_nn[0,:],4*C_dh[0,:]),label = "Identity")
plt.xlabel("n",fontsize = 20)
plt.ylabel("C_nn",fontsize = 20)
plt.legend(loc = 'best',fontsize = 10)
```
